```python
# Sales forecasting model (Linear Regression)
import pandas as pd
from sklearn.linear_model import LinearRegression
import joblib

# Sample training data
df = pd.DataFrame({
    'ad_spend': [50000, 60000, 70000, 80000, 90000],
    'previous_sales': [200000, 220000, 240000, 260000, 280000],
    'revenue': [250000, 270000, 290000, 310000, 330000]
})

# Train model
model = LinearRegression()
model.fit(df[['ad_spend', 'previous_sales']], df['revenue'])

# Save as .pkl (MUST use EXACT name)
joblib.dump(model, 'sales_model.pkl')  # ✅ MUST be "sales_model.pkl" (with underscore)

print("✅ Model saved as sales_model.pkl")
```
